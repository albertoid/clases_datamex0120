```python
# reescalado ([0,1])

def escala(x):
    return (x/100).round(3)
```

```python
conversion=['Hispanic','White', 'Black', 'Native', 
            'Asian', 'Pacific','Poverty',
            'ChildPoverty', 'Service', 'Office']
```

```python
n_columnas=[x+'Rate' for x in conversion]

census[n_columnas]=census[conversion].apply(escala)
census.NativeRate.head()
```

```python
census.head()
```

```python
# normalizacion
import warnings
warnings.simplefilter('ignore')
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

```python
census['TotalPopScaled']=MinMaxScaler().fit_transform(census.TotalPop.values.reshape(-1,1))
```

```python
census.TotalPopScaled.describe()
```

```python
# one-hot, dummies

states=pd.get_dummies(census.State, drop_first=True)
census=pd.concat([census, states], axis=1)
census.head()
```

# Feature Extraction and Engineering

```python
import zipfile
```

```python
zf=zipfile.ZipFile('data/acs2015_county_data.csv.zip')

census=pd.read_csv(zf.open('acs2015_county_data.csv'), index_col='CensusId')

census.head()
```

```python
census.info(memory_usage='deep')
```

```python
# optimizacion de memoria

zf=zipfile.ZipFile('data/mahindra.zip')

ctr=pd.read_csv(zf.open('test.csv'), nrows=1e6)

ctr.head()
```

```python
ctr.info(memory_usage='deep')
```

```python
for e in ctr.select_dtypes('object').columns:
    ctr[e]=ctr[e].astype('category')
    
ctr.info(memory_usage='deep')
```

```python
for e in ctr.select_dtypes('integer').columns:
    ctr[e]=pd.to_numeric(ctr[e], downcast='integer')
    
ctr.info(memory_usage='deep')
```

```python
for e in ctr.select_dtypes('float').columns:
    ctr[e]=pd.to_numeric(ctr[e], downcast='float')
    
ctr.info(memory_usage='deep')
```

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor as DTR
```

```python
features=['Poverty', 'WorkAtHome', 'SelfEmployed', 'Black',
          'Native', 'Asian', 'Pacific', 'Hispanic', 'IncomePerCap']

target='Unemployment'
```

```python
X_train, X_test, y_train, y_test=train_test_split(census[features],
                                                  census[target])
```

```python
modelo=Linreg()

selector=RFE(modelo, n_features_to_select=3, step=1, verbose=1)
selector
```

```python
selector.fit(X_train, y_train)
```

```python
selector.ranking_
```

```python
print ([features[i] for i,e in enumerate(selector.support_) if e])
```
