```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
```

```python
ratings=pd.read_csv('data/movie_ratings.csv').set_index('Movie')
ratings.head()
```

```python
def plot(m1, m2):
    x=ratings.T[m1]
    y=ratings.T[m2]
    
    n=list(ratings.T.index)
    
    fig, ax=plt.subplots()
    ax.scatter(x, y, s=0)
    fig.set_figheight(8)
    fig.set_figwidth(12)
    
    plt.title('Espacio para '+ m1+ ' vs '+ m2, fontsize=20)
    ax.set_xlabel(m1, fontsize=16)
    ax.set_ylabel(m2, fontsize=16)
    
    for i,e in enumerate(n):
        ax.annotate(e, (x[i], y[i]), fontsize=12)
        
```

```python
plot('Venom', 'Incredibles 2')
```

```python
plot('Avengers: Infinity War', 'Deadpool 2')
```

```python
similitud=pd.DataFrame(1/(1+distancias))
similitud
```

```python
# similitud del coseno
import math
import numpy as np

def dot(v1, v2):
    return sum(map(lambda x: x[0]*x[1], zip(v1, v2)))
```

```python
def simil_cos(v1, v2):
    prod=dot(v1, v2)
    len1=math.sqrt(dot(v1, v1))
    len2=math.sqrt(dot(v2, v2))
    return prod/(len1*len2)
```

```python
import scipy.spatial.distance as distance
def plot_v(v):
    origen=(0,0)
    plt.quiver(*origen, v[0,0], v[0,1], color=['r'], scale=60)
    plt.quiver(*origen, v[1,0], v[1,1], color=['b'], scale=60)
    a=v[0]
    b=v[1]
    sim_cos=simil_cos(a, b)
    sim_euc=distance.euclidean(a, b)
    
    tit='Coseno:{}, Euclideo: {}'.format(sim_cos.round(2), sim_euc.round(2))
    plt.title(tit)
```

```python
v=np.array([[10, 10], [20, 10]])
plot_v(v)
```

```python
v=np.array([[10, 0], [0, 10]])
plot_v(v)
```

```python
v=np.array([[10, 10], [0, 20]])
plot_v(v)
```

```python
# generando una recomendacion

tom={'Aquaman':2, 'Avengers: Infinity War':1, 'Black Panther':2,
     'Bohemian Rhapsody':2, 'Deadpool 2':3,
       'Fantastic Beasts: The Crimes of Grindelwald':4, 'Incredibles 2':3,
       'Jurassic World: Fallen Kingdom':2, 'Mission: Impossible – Fallout':5,
       'Venom':5}
```

```python
ratings['Tom']=pd.Series(tom)
ratings.head()
```

```python
distancias=squareform(pdist(ratings.T, 'euclidean'))
similitud=pd.DataFrame(1/(1+distancias), index=ratings.columns, columns=ratings.columns)
similitud
```

```python
similitud.Tom.sort_values(ascending=False)[1:]
```

```python
n_ratings=pd.read_csv('data/movie_ratings2.csv')
n_ratings.head()
```

```python
simil_score=similitud.Tom.sort_values(ascending=False)[1:]
reco=n_ratings.copy()
```

```python
for n,s in dict(simil_score).items():
    reco[n]=reco[n]*s
    
reco['Total']=reco.sum(axis=1)
reco.sort_values('Total', ascending=False)
```
