# Silhouette Score

```python
from sklearn.metrics import silhouette_score
from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from matplotlib.lines import Line2D
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n_sample=1500

X,y=datasets.make_moons(n_samples=n_sample, noise=.05)

X=StandardScaler().fit_transform(X)
```

```python
single=cluster.AgglomerativeClustering(n_clusters=2, linkage='single')

single.fit(X)

y1_pred=single.labels_
```

```python
kmeans=cluster.KMeans(n_clusters=2)

kmeans.fit(X)

y2_pred=kmeans.predict(X)
```

```python
# plot

fig, ax=plt.subplots(1, 2, figsize=(15, 8))
colors=np.array(['blue', 'red'])
legend_e=[Line2D([0], [0], color=c, marker='o', linestyle='', label='Clase {}'.format(i)) for i,c in enumerate(colors)]

ax[0].scatter(X[:,0], X[:,1], color=colors[y1_pred]);
ax[0].legend(handles=legend_e, loc='upper right')

ax[1].scatter(X[:,0], X[:,1], color=colors[y2_pred]);
ax[1].legend(handles=legend_e, loc='upper right')

print ('Modelo 1 Sil Score: {}'.format(silhouette_score(X, y1_pred)))
print ('Modelo 2 Sil Score: {}'.format(silhouette_score(X, y2_pred)))
```

# Elbow

```python
from IPython.display import set_matplotlib_formats
from mpl_toolkits.mplot3d import Axes3D
set_matplotlib_formats('svg')
```

```python
X,y=datasets.make_blobs(n_features=3, centers=2)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:,1], X[:,2], c=y)
```

```python
from sklearn.cluster import KMeans

n_clusters, inercia=zip(*[(k, KMeans(n_clusters=k).fit(X).inertia_) for k in range(1,10)])

fig=plt.figure()
ax=fig.add_subplot()
ax.plot(n_clusters, inercia)
```

```python
from yellowbrick.cluster import KElbowVisualizer

modelo=KMeans()
visual=KElbowVisualizer(modelo, k=(1, 12))
visual.fit(X)
visual.poof();
```

```python
X,y=datasets.make_moons(n_samples=n_sample, noise=.05)

X=StandardScaler().fit_transform(X)
```

```python
inercias=[]

for i in range(1, 10):
    kmeans=KMeans(n_clusters=i).fit(X)
    inercias.append(kmeans.inertia_)
    
plt.plot(range(1, 10), inercias);
```

# Rand Score

```python
from sklearn.metrics import adjusted_rand_score as ars
```

```python
single=cluster.AgglomerativeClustering(n_clusters=2, linkage='single')

single.fit(X)

y1_pred=single.labels_



kmeans=cluster.KMeans(n_clusters=2)

kmeans.fit(X)

y2_pred=kmeans.predict(X)
```

```python
score=ars(y, y1_pred)   # jerarquico
score2=ars(y, y2_pred)  # kmeans

print(score, score2)
```

# Visualizacion de clusters

```python
X , _ = datasets.make_blobs(n_samples=1500, centers=4, cluster_std=1.5)
```

```python
kmeans=KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
y_pred=kmeans.predict(X)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(X[:,0], X[:,1], c=y_pred);
```

```python
centers=kmeans.cluster_centers_
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(X[:,0], X[:,1], c=y_pred);
plt.scatter(centers[:,0], centers[:,1], c='red', s=50, alpha=0.75);
```

```python
# otro 3D

from scipy.spatial.transform import Rotation as R
```

```python
X , _ = datasets.make_blobs(n_samples=1500, centers=3, cluster_std=1.)

X=np.c_[X, 0.5*np.random.randn(X.shape[0], 1)]
```

```python
rot=R.from_euler('y', 45, degrees=True)

X=rot.apply(X)
```

```python
fig=plt.figure(figsize=(12, 8))
ax=fig.gca(projection='3d')

ax.scatter(X[:,0], X[:,1], X[:,2])

ax.set_ylim(-7, 7)
ax.set_xlim(-7, 7)
ax.set_zlim(-7, 7)

ax.view_init(elev=20, azim=65)

plt.show();
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA().fit(X)

pca_x=pca.transform(X)
```

```python
fig=plt.figure(figsize=(12, 8))

ax=[plt.subplot2grid((6, 1), (0,0), rowspan=4), plt.subplot2grid((6, 1), (5,0))]

ax[0].scatter(pca_x[:,0], pca_x[:,1])
ax[0].set_xlabel('PCA 1')
ax[0].set_ylabel('PCA 2')

ax[1].set_title('PCA 3', pad=-5)
ax[1].hlines(1, -7, 10)

y=np.ones(X.shape[0])

ax[1].plot(pca_x[:,2], y, '|', ms='20')
ax[1].axis('off')

plt.show();
```

# Plots t-SNE

```python
from sklearn.manifold import TSNE

X,y=datasets.load_iris(return_X_y=True)
```

```python
fig, ax=plt.subplots(1, 4, figsize=(14, 7))

for i,p in enumerate([5, 30, 50, 100]):
    
    tsne=TSNE(perplexity=p)
    
    x_emb=tsne.fit_transform(X)
    
    ax[i].scatter(x_emb[:,0], x_emb[:,1], c=y)
    ax[i].set_title('Perplexity={}'.format(p))
    
plt.show();
```
